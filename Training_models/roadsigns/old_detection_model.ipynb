### Preparing dataset and labels

```python
#dataset: https://www.kaggle.com/datasets/kasia12345/polish-traffic-signs-dataset
import os

#creating labels for yolo format

for folder_train_val in os.listdir("images"):
    class_id=0
    for folder in os.listdir("images/"+folder_train_val):
        
        for file in os.listdir("images/"+folder_train_val+"/"+folder):
            
            txt_name=os.path.splitext(file)[0]+".txt"
            txt_path=f"labels/{folder_train_val}/{txt_name}"
            print(txt_path)


            
            yolo_label = f"{class_id} 0.5 0.5 1.0 1.0\n"
            print(yolo_label)
            
            with open(txt_path, "w") as f:
                f.write(yolo_label)
        class_id+=1
            
```

```python
import shutil

for folder_train_val in os.listdir("labels"):
    for folder in os.listdir("labels/"+folder_train_val):
        for file in os.listdir("labels/"+folder_train_val+"/"+folder):
            src="labels/"+folder_train_val+"/"+folder+"/"+file
            dst="labels/"+folder_train_val+"/"+file


            shutil.copyfile(src, dst)

            
```
